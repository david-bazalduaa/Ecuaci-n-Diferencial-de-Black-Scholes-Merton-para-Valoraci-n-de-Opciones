## Descripción del Proyecto

Este proyecto tiene como objetivo el uso de ecuaciones diferenciales parciales con técnicas de control y de análisis numérico en un entorno de finanzas. Para esto se plantea el uso de técnicas de control para la valoración de portafolios mediante la ecuación diferencial parcial de Black-Scholes-Merton planteada en [Merton(1973)] y algunas aplicaciones numéricas.

## Fases del Proyecto

### Ecuación diferencial de Black-Scholes-Merton

Aplicar herramientas de teoría de control (función de transferencia, diagrama de bloques, análisis de estabilidad y controlabilidad) al estudio del precio de opciones mediante la ecuación de Black-Scholes-Merton.

La ecuación de Black-Scholes para el precio de un derivado vainilla $V(S, t)$ es:

$$
\frac{\partial V}{\partial t}
+ \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
+ rS \frac{\partial V}{\partial S}
- rV = 0
$$

donde:

- $S(t)$: precio del activo subyacente,  
- $V(S, t)$: valor de la opción,  
- $\sigma$: volatilidad,  
- $r$: tasa libre de riesgo.

---

### Fase 1

Realice el cambio de variable $x = \ln(S)$. Transforme la ecuación de Black-Scholes en una ecuación en términos de $V(x, t)$. Observe que esta transformación permite linealizar el comportamiento de $S(t)$ en escala logarítmica.



---

Cambio de Variable $x = \ln(S)$

1. Derivadas en términos de $x$:
   - $S = e^x \implies \frac{\partial S}{\partial x} = e^x = S$.
   - Por la regla de la cadena:
     $$
     \frac{\partial V}{\partial S} = \frac{\partial V}{\partial x} \cdot \frac{1}{S}, \quad
     \frac{\partial^2 V}{\partial S^2} = -\frac{1}{S^2} \frac{\partial V}{\partial x} + \frac{1}{S^2} \frac{\partial^2 V}{\partial x^2}.
     $$

2. Sustitución en Black-Scholes:

   Ecuación original:
   $$
   \frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} + r S \frac{\partial V}{\partial S} - r V = 0.
   $$
   Sustituyendo:
   $$
   \frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 \left( -\frac{\partial V}{\partial x} + \frac{\partial^2 V}{\partial x^2} \right) + r \frac{\partial V}{\partial x} - r V = 0.
   $$
   Simplificando:
   $$
   \frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 \frac{\partial^2 V}{\partial x^2} + \left( r - \frac{1}{2} \sigma^2 \right) \frac{\partial V}{\partial x} - r V = 0.
   $$

Por lo tanto esta es la ecuación:
$$
\frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 \frac{\partial^2 V}{\partial x^2} + \left( r - \frac{1}{2} \sigma^2 \right) \frac{\partial V}{\partial x} - r V = 0.
$$
